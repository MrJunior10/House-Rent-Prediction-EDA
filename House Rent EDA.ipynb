```python
# Step 1: Load the dataset
df = pd.read_csv('House.csv')
# Step 2: Explore the dataset
# Display the first few rows of the dataset
print(df.head())
# Check the dimensions of the dataset (number of rows, number of columns)
print(df.shape)
# Check the data types of each column
print(df.dtypes)
# Step 3: Handle missing values (if any)
# Check for missing values in the dataset
print(df.isnull().sum())
# Handle missing values (if required)
# For example, to fill missing values with the mean of the column
# df['column_name'].fillna(df['column_name'].mean(), inplace=True)
# Step 4: Perform basic statistical analysis
# Generate descriptive statistics for numerical columns
print(df.describe())
# Calculate the correlation matrix for numerical columns
correlation_matrix = df[['Rent', 'Size', 'Bathroom']].corr()
print(correlation_matrix)
# Step 5: Data Visualization
# Histogram of a numerical column
plt.hist(df['Rent'], bins=10)
plt.xlabel('Rent')
plt.ylabel('Frequency')
plt.title('Distribution of Rent')
plt.show()
# Box plot of a numerical column
plt.boxplot(df['Size'])
plt.xlabel('Size')
plt.title('Box Plot of Size')
plt.show()
# Scatter plot of two numerical columns
plt.scatter(df['Rent'], df['Size'])
plt.xlabel('Rent')
plt.ylabel('Size')
plt.title('Scatter Plot of Rent vs Size')
plt.show()
```

```python
# Heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Rent', 'Size', 'Bathroom']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot matrix
numerical_cols = ['Rent', 'Size', 'Bathroom']
sns.pairplot(df[numerical_cols])
plt.title('Scatter Plot Matrix')
plt.show()
```

```python
# Box plot grouped by a categorical variable
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='City', y='Rent')
plt.xlabel('City')
plt.ylabel('Rent')
plt.title('Rent Distribution by City')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Furnishing Status', y='Rent')
plt.xlabel('Furnishing Status')
plt.ylabel('Rent')
plt.title('Rent Distribution by Furnishing Status')
plt.show()
```

```python
# Pairwise scatter plot with hue
numerical_cols = ['Rent', 'Size', 'Bathroom']
sns.pairplot(df, vars=numerical_cols, hue='Furnishing Status', plot_kws={'alpha': 0.5})
plt.title('Pairwise Scatter Plot with Hue')
plt.show()
```
